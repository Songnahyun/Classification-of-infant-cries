```python
#White Noise
def adding_white_noise(y,sr=22050,noise_rate=0.005 ):
  wn = np.random.randn(len(y))
  y_wn = y+ noise_rate*wn
  librosa.display.waveplot(y_wn)
  for i in range(1,6):
    sf.write(f'/content/drive/MyDrive/baby/white_noise(투정{i}).wav', y, sr,format='WAV') # 저장
  print('White Noise 저장 성공')
  return y  
```

```python
#Shifting
def shifting_sound(y, sr=22050, roll_rate=0.1):
    # [1, 2, 3, 4] 를 [4, 1, 2, 3]으로 만들어 주는 것
    data_roll = np.roll(y, int(len(y) * roll_rate))
    librosa.display.waveplot(data_roll)
    for i in range(1,6):
      sf.write(f'/content/drive/MyDrive/baby/shifting_sound(투정{i}).wav', y, sr, format = 'wav')
    print('shifting_sound 저장 성공')
    
    return y
```

```python
shifting_sound(y)
```

```python
#Stretching
def stretch_sound(y, sr=22050, rate=0.8):
    # stretch 해주는 것 테이프 늘어진 것처럼 들린다.
    stretch_data = librosa.effects.time_stretch(y, rate)
    librosa.display.waveplot(stretch_data)
    for i in range(1,6):
      sf.write(f'/content/drive/MyDrive/baby/stretch_data(투정{i}).wav', stretch_data, sr, format= 'wav')
    print('stretch_data 저장 성공')
    
    return y
```

```python
stretch_sound(y)
```

```python
#Minus
def minus_sound(y, sr=22050):
    # 위상을 뒤집는 것으로서 원래 소리와 똑같이 들린다.
    temp_numpy = (-1)*y
    librosa.display.waveplot(temp_numpy)
    for i in range(1,6):
      sf.write(f'/content/drive/MyDrive/baby/minus_data(투정{i}).wav', temp_numpy, sr, format='wav')
    print('minus_data 저장 성공')

    return y
```

```python
minus_sound(y)
```
